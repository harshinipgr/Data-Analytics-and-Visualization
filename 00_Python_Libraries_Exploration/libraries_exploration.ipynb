```python
# Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
import plotly.express as px
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

print("All libraries imported successfully!")
```

# 1. NumPy

NumPy is used for numerical computing and efficient array operations.

```python
arr = np.array([10,20,30,40,50])

print("Array:", arr)
print("Mean:", np.mean(arr))
print("Maximum:", np.max(arr))
print("Minimum:", np.min(arr))
print("Sum:", np.sum(arr))
```

# 2. Pandas

Pandas is used for data manipulation and analysis.

```python
df = pd.DataFrame({
    "Student":["Alice","Bob","Charlie","David"],
    "Marks":[85,92,78,95]
})

df
```

```python
print(df.describe())
```

# 3. Matplotlib

Matplotlib is used for creating basic charts.

```python
plt.figure(figsize=(6,4))

plt.bar(df["Student"], df["Marks"])

plt.title("Student Marks")

plt.xlabel("Students")

plt.ylabel("Marks")

plt.show()
```

# 4. Seaborn

Seaborn provides beautiful statistical visualizations.

```python
plt.figure(figsize=(6,4))

sns.barplot(data=df, x="Student", y="Marks")

plt.title("Student Marks using Seaborn")

plt.show()
```

# 5. Plotly

Plotly creates interactive charts.

```python
fig = px.bar(
    df,
    x="Student",
    y="Marks",
    title="Interactive Plotly Chart"
)

fig.show()
```

# 6. Bokeh

Bokeh is used for interactive visualizations.

```python
p = figure(
    title="Student Marks",
    x_range=df["Student"].tolist(),
    width=600,
    height=350
)

p.vbar(
    x=df["Student"],
    top=df["Marks"],
    width=0.5
)

show(p)
```

# 7. SciPy

SciPy provides scientific and statistical functions.

```python
print("Mean:", stats.tmean(arr))
print("Variance:", stats.tvar(arr))
print("Standard Deviation:", stats.tstd(arr))
```

# 8. Statsmodels

Statsmodels is used for statistical modeling.

```python
X = [1,2,3,4,5]
Y = [2,4,5,4,5]

X = sm.add_constant(X)

model = sm.OLS(Y, X).fit()

print(model.summary())
```
